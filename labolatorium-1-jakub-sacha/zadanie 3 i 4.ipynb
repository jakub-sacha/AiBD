Zadanie to polegało na zdefiniowaniu funkcji ***x^2+5*** do czego użyłem funkcji ***lambda***, a następnie sporządzeniu wykresu
dla 3 zakresów:

* -1 < x < 1
* -6 < x < 6
* 0 < x < 5

Zostały dodane również etykiety osi, tytuły wykresów i legendy. Do wizualizacji wyników został wykorzystany pakiet ***Matplotlib***.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: x**2 + 5
first_range = np.linspace(-1, 1, 100)
second_range = np.linspace(-6, 6, 100)
third_range = np.linspace(0, 5, 100)

plt.plot(first_range, f(first_range), "r-", label="Pierwszy zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres 1')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(second_range, f(second_range), "g-", label="Drugi zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres 2')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(third_range, f(third_range), "y-", label="Trzeci zakres")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Wykres 3')
plt.grid()
plt.legend()
plt.show()
```
